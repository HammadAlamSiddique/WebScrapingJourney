# **Exercise: Scrape job postings details from Indeed**
#### You have to scrape by following steps:
1- Input job title into input box\
2- Get link, title, company, salary, date, location\
3- Do this for every page until no jobs left

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import pandas as pd
import time
import requests
```

```python
driver = webdriver.Chrome('C:/Users/Hammad/Downloads/Compressed/chromedriver.exe')
driver.get('https://pk.indeed.com/')
time.sleep(3)
```

---
### Enter job title and location here:

```python
job_title = 'data entry'
location = 'karachi'
```

---

```python
job_title_box = driver.find_element(By.XPATH, '//*[@id="text-input-what"]')
job_title_box.send_keys(job_title)
```

```python
city_box = driver.find_element(By.XPATH, '//*[@id="text-input-where"]')
city_box.send_keys(location)
```

```python
find_button = driver.find_element(By.XPATH, '//*[@id="jobsearch"]/button')
find_button.click()
time.sleep(4)
```

```python
df = pd.DataFrame({'Link':[''], 'Job Title':[''], 'Company Name':[''], 'Location':[''], 'Salary':[''], 'Date':[''] })
```

```python
while True:
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    postings = soup.find_all('div', {'class':'slider_container css-g7s71f eu4oa1w0'})

    for post in postings:
        link = post.find('a', {'class':'jcs-JobTitle css-jspxzf eu4oa1w0'}).get('href')
        link_full = 'https://pk.indeed.com'+link
        job_title = post.find('div', {'class':'css-1m4cuuf e37uo190'}).text
        company_name = post.find('span', {'class':'companyName'}).text
        date = post.find('span', {'class':'date'}).text

        try:
            location = post.find('div', {'class':'companyLocation'}).text
        except:
            location = 'N/A'
        
        try:
            salary = post.find('div', {'class':'metadata salary-snippet-container'}).text
        except:
            salary = 'N/A'


        df = df.append({'Link':link_full, 'Job Title':job_title, 'Company Name':company_name, 'Location':location, 'Salary':salary, 'Date':date},
                                    ignore_index=True )

    try:
        # isme mene class b use ki thi lekin attrs ne kaam kardia 
        next_page = soup.find('a', attrs = {'data-testid':'pagination-page-next'}).get('href')
        next_page_full = 'https://pk.indeed.com'+next_page
        driver.get(next_page_full)
    except:
        break
```

```python
df.to_csv('indeed-jobs.csv')
```
