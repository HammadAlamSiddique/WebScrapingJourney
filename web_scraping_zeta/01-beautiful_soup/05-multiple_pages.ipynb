```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.airbnb.com/s/Europe/homes?tab_id=home_tab&refinement_paths%5B%5D=%2Fhomes&flexible_trip_lengths%5B%5D=one_week&price_filter_input_type=0&price_filter_num_nights=5&place_id=ChIJhdqtz4aI7UYRefD8s-aZ73I&date_picker_type=calendar&checkin=2022-12-25&checkout=2022-12-31&source=structured_search_input_header&search_type=filter_change"
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
df = pd.DataFrame({'Links':[''], 'Title':[''], 'Details':[''], 'Rating':[''], 'Price':[''] })
```

```python
counter = 1
while counter <= 15:

    postings = soup.find_all('div', class_='c4mnd7m dir dir-ltr')
    for post in postings:
        link = post.find('a', class_='bn2bl2p dir dir-ltr').get('href')
        link_full = "https://www.airbnb.com"+link
        title = post.find('div', class_='t1jojoys dir dir-ltr').text

        try:
            price = post.find('span', class_='a8jt5op dir dir-ltr').text
        except:
            price = 'N/A'

        try:    
            rating = post.find('span', class_='r1dxllyb dir dir-ltr').text
        except:
            rating = 'N/A'
        
        try:
            details = post.find('div', class_='nquyp1l s1cjsi4j dir dir-ltr').text
        except:
            details = 'N/A'

            
        df = df.append({'Links':link_full, 'Title':title, 'Details':details, 'Rating':rating, 'Price':price }, ignore_index=True)


    next_page = soup.find('a', {'aria-label':'Next'}).get('href')
    next_page_full = "https://www.airbnb.com"+next_page


    url = next_page_full
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    
    counter += 1
```

```python
df.to_excel('airbnb.xlsx', index=False)
```
