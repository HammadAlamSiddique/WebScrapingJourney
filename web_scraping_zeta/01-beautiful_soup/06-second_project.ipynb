# Project 02: Scraping 'Car Pages' website
### You have to scrape the following:
- get he link of each posting
- get the name of each car
- get the pice of each car
- get the color of each car
- pull all that data into a table
- do this for the first 10 - 15 pages on the site

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.carpages.ca/used-cars/search/?fueltype_id%5B0%5D=3&fueltype_id%5B1%5D=7"
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
df = pd.DataFrame({'Title':[''], 'Price':[''], 'Color':[''], 'Link':[''] })
```

```python
counter = 0
```

```python
while counter < 10:

    postings = soup.find_all('div', class_='media soft push-none rule')

    for post in postings:
        link = post.find('a', class_='media__img media__img--thumb').get('href')
        link_full = "https://www.carpages.ca/"+link
        title = post.find('h4', class_='hN').get_text(strip=True)
        price = post.find('strong', class_='delta').get_text(strip=True)
        color = post.find_all('div', class_='grey l-column l-column--small-6 l-column--medium-4')[1].get_text(strip=True)
        df = df.append({'Title':title, 'Price':price, 'Color':color, 'Link':link_full }, ignore_index=True)


    next_page = soup.find('a', class_='nextprev').get('href')
    next_page_full = "https://www.carpages.ca/"+next_page
    next_page_full

    url = next_page_full
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    counter += 1
```

```python
df.to_excel('car-pages.xlsx')
```
